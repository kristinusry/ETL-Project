```python
# Dependencies
import pandas as pd
```

# Folsom Weather Data from OpenWeather

```python
# Import Folsom Lake historical weather export
FOLweather_df = "data/d6b7778d4112138039dabc3de94f08ab.csv"
FOLweather_df = pd.read_csv(FOLweather_df)
FOLweather_df['dt'] = pd.to_datetime(FOLweather_df['dt'], unit='s') #convert unix to datetime
FOLweather_df = FOLweather_df.rename(columns = {'dt':'date_time'}) #rename 'dt' column
FOLweather_df.head()
```

```python
#Get all column names
list(FOLweather_df.columns)
```

```python
#Create dataframe with only desired columns
FOLweather_clean_df = FOLweather_df[["date_time","temp_max","humidity","wind_speed","weather_main","clouds_all"]]
FOLweather_clean_df
```

# Folsom Lake Elevation Data from CDEC

```python
# Import Folsom Lake elevation
FOLelevation_df = "data/FOL_6.csv"
FOLelevation_df = pd.read_csv(FOLelevation_df)
FOLelevation_df['DATE TIME'] = pd.to_datetime(FOLelevation_df['DATE TIME']) #convert string to datetime
FOLelevation_df = FOLelevation_df.rename(columns = {'DATE TIME':'date_time','VALUE':'elevation'}) #rename columns
FOLelevation_df.head()
```

```python
FOLelevation_clean_df = FOLelevation_df[["date_time", "elevation"]]
FOLelevation_clean_df
```

# Join Folsom Lake Weather and Elevation

```python
FOL_weather_elevation_df = pd.merge(FOLweather_clean_df, FOLelevation_clean_df, on=['date_time'], how="inner")
FOL_weather_elevation_df
```

```python
#Check if any columns have NaN values 
FOL_weather_elevation_df.isna().any()
```

```python
#Appears that CDEC is missing some elevation data
#View a sample of the Nan values
FOL_weather_elevation_df[FOL_weather_elevation_df['elevation'].isnull()]
```

```python
#Drop the rows with NaN elevation values
FOL_weather_elevation_df = FOL_weather_elevation_df.dropna()
FOL_weather_elevation_df
```

```python
#Verify that no columns have NaN values 
FOL_weather_elevation_df.isna().any()
```
